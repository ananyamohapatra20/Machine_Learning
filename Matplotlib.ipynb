Useful for making plots

```python
#Import matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)#sine value
z = np.cos(x)#cosine value
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting

```python
#Sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#Adding title x-axis & y-axis labels.
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine wave')
plt.show()
```

```python
#Parabola
m = np.linspace(-10,10,20)
n = m**2
plt.plot(m,n)
plt.show()
```

```python
plt.plot(m,n, 'r+')
plt.show()
```

```python
plt.plot(m,n, 'g*')
plt.show()
```

```python
plt.plot(m,n, 'bx')
plt.show()
```

```python
s = np.linspace(-5,5,50)
plt.plot(s,np.sin(s),'g-')
plt.plot(s,np.cos(s),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
sub=['English','Hindi','Sanskrit','Telgu']
stu=[100,50,43,63]
ax.bar(sub,stu)
plt.title('Student Report')
plt.xlabel('Subjects')
plt.ylabel('No of Students')
plt.show()
```

Pie Chart

```python
fig1=plt.figure()
ax = fig1.add_axes([0,0,1,1])
sub=['English','Hindi','Sanskrit','Telgu']
stu=[100,50,43,63]
ax.pie(stu,labels=sub, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x= np.linspace(0,10,30)
y= np.sin(x)
z= np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z= 20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
