Manipulating a Dataframe

```python
# add a column in the dataframe
sample=pd.read_csv("/content/dia.csv")
sample.head()
```

```python
sample['New_Col']=sample['Glucose']+sample['BloodPressure'] #adding column in the DataFrame
```

```python
sample.head()
```

```python
#Remove a row
sample.drop(index=0,axis=0)
```

```python
#Remove a column
sample1=sample.drop(columns='Unnamed: 0',axis=1)
```

```python
sample1.drop(columns='New_Col')
```

```python
#Locating Row using the Index Value
sample.iloc[2]
```

```python
#locating particular column
print(sample1.iloc[:,0])#First col
print(sample1.iloc[:,1])#Second col
print(sample1.iloc[:,3])#Third col
print(sample1.iloc[:,-1])#Last Column
```
