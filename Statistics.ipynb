```python
iqr=q3-q1
lower_fence=q1-(1.5*iqr)
higher_fence=q3+(1.5*iqr)
print(lower_fence,higher_fence)
```

<h3>Measure of Dispersion</h3>
1.Variance<br>
2.Standard Deviation

```python
statistics.variance(ages) # Sample variance uses divided by n-1 formula
```

```python
np.var(ages,axis=0) # Population variance uses divided by n formula
```

```python
def Variance(age):
  n=len(ages)
  mean=sum(age)/n
  deviation=[(x - mean) ** 2 for x in age]
  variance=sum(deviation)/n #For Population
  return variance
```

```python
print(Variance(ages))
```

```python
def Var(age,dof=0):
  n=len(ages)
  mean=sum(age)/n
  deviation=[(x - mean) ** 2 for x in age]
  variance=sum(deviation)/(n-dof) #For Sample dof=1
  return variance
```

```python
print(Var(ages,dof=1))
```

```python
statistics.pvariance(ages) #population variance
```

```python
import math
math.sqrt(statistics.pvariance(ages)) #Standard deviation
```

<h1>Check Whether a distribution is Normal Distribution</h1>

If we want to check whether feature is gaussian or normal distributed<br>
We have to create Q-Q Plot<br>
QQ plot (quantilequantile plot) is a probability plot, a graphical method for comparing two probability distributions by plotting their quantiles against each other.

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist='norm',plot=pylab)
  plt.show()
```

```python
dta=np.random.normal(0.2,0.5,1000)
print(plot_data(dta))
```

```python
print(plot_data(data['sepal_width']))
```

To check for Log Normal Distribution

```python
def plot(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample),dist='norm',plot=pylab)
  plt.show()
```

```python
samp=np.random.lognormal(3,1.5,100)
print(plot(samp))
```
